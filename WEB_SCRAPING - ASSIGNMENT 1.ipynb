## Question 1

Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
def question_1(url):
    
    result = requests.get(url)  #making request
    soup = BeautifulSoup(result.content) #Bringing to Correct html Format
    
    h1_h2_h3_tags = []     #creating a empty list
    for i in range(1,4):   #for h1,h2,h3 tags
        header_tag = 'h{}'.format(i)
        headers = soup.find_all(header_tag)
        for j in headers:
            h1_h2_h3_tags.append(j.text.replace('\n',''))#Lets append all the header tags(h1,h2,h3) to to the list.
    return h1_h2_h3_tags   #returning the list
```

```python
#calling the function to display all header tags in the list
question_1('https://en.wikipedia.org/wiki/Main_Page')
```

## Question 2

Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name,IMDB rating, Year of release) 
#save it in form of a CSV file.

```python
def question_2(url):
    result = requests.get(url)   #making request
    soup = BeautifulSoup(result.content)    #Bringing to Correct html Format using beautiful soup
    
    
    Movie_dataframe = pd.DataFrame({})   #creating a empty dataframe
    
    Movie_list = []     #creating a empty list
    Movies = soup.find_all('td',class_= 'titleColumn')   #finding the tag associated with particular class
    for i in range(len(Movies[0:100])):   #we need only 100 movies
        name = Movies[i].select('a')
        Movie_list.append(name[0].text)  #appending it to the list
    
    Realease_year_list = []
    years = soup.find_all('span',class_= 'secondaryInfo')
    for year in years[0:100]:
        i = year.text.replace('(','')
        Realease_year_list.append(i.replace(')',''))
    
    Rating_list = []
    Movie_ratings = soup.find_all('td',class_= 'ratingColumn imdbRating')
    for rating in Movie_ratings[0:100]:
        Rating_list.append(rating.text.replace('\n',''))
    
    Movie_dataframe['Rank'] = list(range(1,101))    #top 100 movies
    Movie_dataframe['Name'] = Movie_list            #Creating a column for the particular list
    Movie_dataframe['Year_of_Release'] = Realease_year_list
    Movie_dataframe['Ratings'] = Rating_list
    
    
    Movie_dataframe.to_csv('Top_Rated_Movie_worldwide.csv',index = False) #saving it in form of a CSV file

    
    
    return Movie_dataframe  #returning the Dataframe as result
```

```python
#calling the function to display the dataframe as result
question_2('https://www.imdb.com/chart/top/')
```

## Question 4

Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’


```python
def question_4(url):
    result = requests.get(url)  #making request
    soup = BeautifulSoup(result.content)  #Bringing to Correct html Format using beautiful soup
    
    
    Books_dataframe = pd.DataFrame({})   #creating a empty dataframe
    
    Books_list = []  #creating a empty list
    Author_list = []
    Genre_list = []
    Review_list = []

    Books = soup.find_all('h4',class_= 'italic')   #finding the tag associated with particular class
    Authors = soup.find_all('p',class_ ='sans bold')
    Genres = soup.find_all('p',class_ ='genre-links hidden-phone')
    Reviews = soup.find_all('p',class_ ='excerpt')
    
    for book in Books:
        name = book.text.replace('★','') #replacing the unwanted characters with null
        name = name.replace('\n','')
        Books_list.append(name)   #appending it to the list

    for author in Authors:
        Author_list.append(author.text.replace('\n',''))
    
    for genre in Genres:
        Genre_list.append(genre.text.replace('\n',''))
    
    for reviwe in Reviews:
        Review_list.append(reviwe.text.replace('\n',''))
    
    Books_dataframe['Book Name'] = Books_list  #Creating a column for the particular list
    Books_dataframe['Author'] = Author_list
    Books_dataframe['Genre'] = Genre_list
    Books_dataframe['Review'] = Review_list
    
    return Books_dataframe  #returning the Dataframe as result
```

```python
#calling the function to display the dataframe as result
question_4('https://bookpage.com/reviews')
```

### Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’.


i) Top 10 ODI teams in men’s cricket along with the records for matches, points and
rating. 

ii) Top 10 ODI Batsmen in men along with the records of their team and rating. 

iii)Top 10 ODI bowlers along with the records of their team and rating.

```python
#Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
def question5_part1(url):
    
    result = requests.get(url)  #making request
    
    soup = BeautifulSoup(result.content)  #Bringing to Correct html Format using beautiful soup
    
    ODI_table = soup.select('td',class_ = 'table-body')  #finding the tag associated with particular class
    
    top_ten_table = ODI_table[:50]   #since we need only top 10 
    
    top_ten_teams = pd.DataFrame({})   #creating a empty dataframe
    
    Column_names = ['Position','Team','Mathces','Points','Ratings']  #creating the columns required in dataframe in form of a list
    
    
    for i in range(len(Column_names)):  
        temp_list = []   #creating a temperory list for every columns
        for j in top_ten_table[i::5]:    #since we have 50 objects in list, using skip with 5
            temp_list.append(j.text.replace('\n',''))
        
        top_ten_teams[Column_names[i]] = temp_list  #appending it to the list

    
    return top_ten_teams  #returning the Dataframe as result
```

## Question 7

```python
#7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in.
#The scraped data should include Product Name, Price, Image URL and Average Rating.
def question7(urls):
    
    x = 0

    while x == 0:  #creating while loop because sometimes request will be empty
        result = requests.get(urls)    #making request
        soup = BeautifulSoup(result.content)   #Bringing to Correct html Format using beautiful soup
        Raw_data = soup.find_all('div',class_= 'a-section a-spacing-medium')  #finding the tag associated with particular class
        x = len(Raw_data)   #if any data is retrived , then only while loop will break
    
    
    
    cost_list = []   #creating a empty list
    name_list = []
    image_list = []
    rating_list = []
    for j in range(len(Raw_data)):
        name = Raw_data[j].find('span',class_ = 'a-size-medium a-color-base a-text-normal').text   #finding the tag associated with particular class
        rating = Raw_data[j].find('span',class_ = 'a-icon-alt').text
        image = Raw_data[j].select('img',class_='s-image')[0]['src']
        cost = Raw_data[j].find('span',class_ ='a-price-whole')           
        if cost is None:   #handing nonetype error
            continue
        else:
            cost_int = int(cost.text.replace(',',''))  #converting to integer
            if cost_int <= 20000:
                cost_list.append(cost_int)  #appending it to the list
                name_list.append(name)
                rating_list.append(rating)
                image_list.append(image)
 
    mobile_dataframe  = pd.DataFrame({})     #creating a empty dataframe

    mobile_dataframe['Mobile Name'] = name_list  #creating the columns required in dataframe in form of a list
    mobile_dataframe['Cost'] = cost_list
    mobile_dataframe['Ratings'] = rating_list
    mobile_dataframe['Image url'] = image_list
    
    mobile_dataframe.to_csv('Mobile_under_20,000.csv')  #saving it in form of a CSV file

    return mobile_dataframe  #returning the Dataframe as result
    
```

```python
#calling the function to display the dataframe as result
amazon_url = 'https://www.amazon.in/s?k=mobiles&rh=n%3A1389401031&ref=nb_sb_noss'
question7(amazon_url)
```

## Question 8

```python
def question8(url):
    
    result = requests.get(url)  #making request
    
    soup = BeautifulSoup(result.content)  #Bringing to Correct html Format using beautiful soup
    
    weather_dataframe = pd.DataFrame({})   #creating a empty dataframe
    
    Raw_data = soup.find_all('div',class_ ='tombstone-container')  #finding the tag associated with particular class
    
    Periods_list = []  #creating a empty list
    for i in Raw_data:
        z = i.select('p')
        Periods_list.append(z[0].text)   #appending it to the list
    
    Short_des_list = []
    for i in Raw_data:
        y = i.select('p')
        Short_des_list.append(y[2].text)
    
    Temperature_list = []
    for i in Raw_data:
        z = i.select('p')
        Temperature_list.append(z[3].text)
    
    Description_list = []
    Descriptions = soup.find_all('div',class_ ='col-sm-10 forecast-text')
    for description in Descriptions[:9]:
        Description_list.append(description.text)
    
    weather_dataframe['Period'] = Periods_list   #Creating a column for the particular list
    weather_dataframe['Short Description'] = Short_des_list
    weather_dataframe['Temperature'] = Temperature_list
    weather_dataframe['Description'] = Description_list
    
    return weather_dataframe  #returning the Dataframe as result
```

```python
#calling the function to display the dataframe as result
question8('https://forecast.weather.gov/MapClick.php?lat=37.7771&lon=-122.4196#.YJG6I7UzbIU')
```
