```python
#lets import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
import re
pd.set_option('display.max_row',None)
```

## Question 1

```python
def question1(url):
    #activate chrome browser
    driver= webdriver.Chrome("chromedriver.exe")
    #opening the amazon.com
    driver.get(url)
    
    #lest set the delay
    delay =15
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME ,'nav-search-field ')))


    #finding the element for product search bar
    input_bar = driver.find_element_by_xpath("//div[@class='nav-search-field ']/input")

    #write on search bar
    user_input = input('Enter the Product name : ')
    input_bar.send_keys(user_input)
    
    #finding the search button and clicking on it
    search_key = driver.find_element_by_id('nav-search-submit-button')
    search_key.click()
    
    #getting the url of the searched product
    question1_url = driver.current_url
    
    driver.close()
    
    return question1_url #returning url as result
```

## Question 2

```python
def question2(url):

    
    product_url = question1(url)  #calling question1 inside question2
    
        
    #activate chrome browser
    driver= webdriver.Chrome("chromedriver.exe")  #activating the chrome driver
    
    driver.get(product_url)   #url returned from question 1
    

    #creating the empty lists 
    product_name_list = []
    brand_name_list = []
    ratings_list = []
    no_of_rating_list = []
    price_list = []
    expected_delivery_list = []
    exchange_return_list = []
    availability_list= []
    other_details_list = []
    guitars_links_list = []  

    for k in range(3):  #to get the product links from 3  pages

        delay =15 #setting dealy
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME ,'a-last')))

        #product links
        links = driver.find_elements_by_xpath("//h2[@class ='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a")

        for j in links:
            guitars_links_list.append(j.get_attribute('href'))  #appending the links to the list

        #for not clicking next button after page 3
        if k < 2 :
            next_button = driver.find_element_by_xpath("//li[@class= 'a-last']")
            next_button.click()
    
    #looping through all the product links obtained from 3 pages
    
    for i in guitars_links_list:

        driver.get(i)

        delay =15 #setting delay for review, beacuse it loads at last
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.ID ,'reviewsMedley')))

        #time.sleep(2)
        
        #using try catch for all the columns
        
        try:
            product_name = driver.find_element_by_id("productTitle")
            product_name_list.append(product_name.text)
        except NoSuchElementException:
            product_name_list.append('-')

        try:
            brand_name = driver.find_element_by_xpath("//table[@class = 'a-normal a-spacing-micro']/tbody/tr[1]/td[2]")
            brand_name_list.append(brand_name.text)
        except NoSuchElementException:
            brand_name_list.append('-')

        try:
            ratings = driver.find_element_by_xpath("//span[@class = 'a-size-base a-nowrap']/span[1]")
            ratings_list.append(ratings.text)
        except NoSuchElementException:
            ratings_list.append('-')

        try:
            no_of_rating = driver.find_element_by_xpath("//span[@class = 'a-size-base a-color-secondary']")
            no_of_rating_list.append(no_of_rating.text)
        except NoSuchElementException:
            no_of_rating_list.append('-')
        
        #for price column, there is two criteria 
        #price
        #deal of the day
        try:
            price = driver.find_element_by_id('priceblock_ourprice')
            price_list.append(price.text)
        except NoSuchElementException:
            try:
                price = driver.find_element_by_id('priceblock_dealprice')
                price_list.append(price.text)
            except NoSuchElementException:
                price_list.append('-')

        try:
            availability = driver.find_element_by_id("availability")
            availability_list.append(availability.text)
        except NoSuchElementException:
            availability_list.append('-')

        try:
            other_details = driver.find_element_by_id('feature-bullets')
            other_details_list.append(other_details.text)
        except NoSuchElementException:
            other_details_list.append('-')

        try:
            expected_delivery = driver.find_element_by_xpath("//div[@id = 'ddmDeliveryMessage']/b")
            expected_delivery_list.append(expected_delivery.text)
        except NoSuchElementException:
            expected_delivery_list.append('Pincode not entered / Not available')

        try:
            exchange_return = driver.find_element_by_xpath("//div[@id = 'icon-farm-container']/div/div")
            exchange_return_list.append(exchange_return.text)
        except NoSuchElementException:
            exchange_return_list.append('-')

    df_guitar = pd.DataFrame({}) #creating the empty dataframe
    df_guitar['Brand'] = brand_name_list
    df_guitar['Name'] = product_name_list  #creating the columns assciated with list
    df_guitar['Price'] = price_list
    df_guitar['Ratings'] = ratings_list
    df_guitar['Number of Ratings'] = no_of_rating_list
    df_guitar['Availabity'] =availability_list
    df_guitar['Return/Exchange'] = exchange_return_list
    df_guitar['Expected Delivery'] = expected_delivery_list
    df_guitar['Other Details'] =other_details_list
    df_guitar['URL'] = guitars_links_list
    
    driver.close()#closing the driver
    
    return df_guitar  #returing the dataframe as the result
```

## Question 3

```python
#activate chrome browser
def Question3(url):
    
    #activating the browser
    driver= webdriver.Chrome("chromedriver.exe")
    
    #creating the list of itmes we need
    Mylist = ['Fruits','Cars', 'Machine Learning']
    
    
    #creating the empty dataframe
    fruits_cars_ml_df = pd.DataFrame({})


    #looping through our created list
    for i in Mylist:
        
        #creating a temp list instead of creating indivial list for all
        temp_list = []

        #making request to images.google.com
        driver.get(url)
        
        #input bar
        input_bar = driver.find_element_by_xpath("//input[@class = 'gLFyf gsfi']")
    
        #sending our list items one by one
        input_bar.send_keys(i)
        
        #clicking on search button
        search_button = driver.find_element_by_xpath("//button[@class = 'Tg7LZd']")
        search_button.click()
        
        #creating a while loop to make we get 100 images
        


        x = 0
        while x < 100:
            
            temp_list.clear() #clearing the list every time, so we don't get duplicates
            
            time.sleep(3)
            image_elements = driver.find_elements_by_xpath("//img[@class = 'rg_i Q4LuWd']") #getting suitable tag

            for j in image_elements:
                img_url = j.get_attribute('src')
                if img_url is not None:
                    temp_list.append(img_url) #appending image links to the list if element is not null
            x = len(temp_list)
            driver.execute_script("window.scrollBy(0,document.body.scrollHeight)")
            
            
        fruits_cars_ml_df[i] = temp_list[:100]   #creating the column for the dataframe
    
    driver.close()
    
    return fruits_cars_ml_df  #returning the Dataframe as result
```

```python
#I am using Regular expressions in this question, since regex is very powerful and fast

def question4(url):
    #activate chrome browser
    driver= webdriver.Chrome("chromedriver.exe")

    #opening the flipcart.com
    driver.get(url)


    #we can skip the login option by closing the popup
    #because i am getting loh=gin option everytime
    #if you are getting login popup, please comment out the below three lines
    delay = 15
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME , '_1-pxlW'))) 
    
    #lets skip the login 
    skip_login = driver.find_element_by_xpath("//button[@class ='_2KpZ6l _2doB4z']")
    skip_login.click()

    #setting delay
    delay = 15
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME , '_3OO5Xc'))) 
    
    #finding the element for product search bar
    product_search = driver.find_element_by_xpath("//div[@class = '_3OO5Xc']/input")
    #write on search bar
    product_search.send_keys("Pixel A4")

    #finding the search button with xpath
    search =driver.find_element_by_xpath("//button[@class = 'L0Z3Pu']") 
    search.click()  #clicking the button


    #lets set the delay based on class
    delay = 15
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME , '_4rR01T'))) 
    
    #creating the empty list
    phone_names_list = []
    phone_colour_list = []
    ram_list = []
    rom_list = []
    front_cam_list =[]
    back_cam_list = []
    battery_list = []
    display_list = []
    processor_list = []
    price_list = []
    phone_desc_list = []
    phone_url_list = []
    
    #getting all above 12 columns in the first page only without looping into every page using regex .
    
    #price column using regex
    price = driver.find_elements_by_xpath("//div[@class = '_30jeq3 _1_WHN1']")
    for i in price:
        price_list.append(i.text)

    
    #lets extract string called phone description from every phone in first page
    phone_desc = driver.find_elements_by_xpath("//div[@class= '_4rR01T']")
    for i in phone_desc:
        phone_desc_list.append(i.text)
    
    #from the string extracted from every phone, lets use regex to extract the required data for ecah columns
    for i in phone_desc_list:
        try:
            colour = re.search(r'\D[\w\s]+,',i).group().replace('(','')
            phone_color = colour.replace(',','')
            phone_colour_list.append(phone_color)
        except AttributeError:
            phone_colour_list.append('-')

        try :
            name = re.search(r'[\d\w\s]+\W',i).group().replace('(','')
            phone_names_list.append(name)
        except AttributeError:
            phone_names_list.append('-')

    phone_url = driver.find_elements_by_xpath("//a[@class = '_1fQZEK']")
    
    for i in phone_url:
        phone_url_list.append(i.get_attribute('href'))


    full_desc = driver.find_elements_by_xpath("//ul[@class =  '_1xgFaf']")
    for i in full_desc:

        mystr = i.text

        try:
            ram = re.search(r'([\d]+\s)GB\sRAM',mystr)
            ram_list.append(ram.group())
        except AttributeError:
            ram_list.append('-')

        try:
            rom = re.search(r'([\d]+\s)GB\sROM',mystr)
            rom_list.append(rom.group())
        except AttributeError:
            rom_list.append('-')

        try:
            front_camera = re.search(r'\d\w{2}\sFront\sCamera',mystr)
            front_cam_list.append(front_camera.group())
        except AttributeError:
            front_cam_list.append('-')

        try:
            rear_camera = re.search(r'\d\w{2}\sRear\sCamera',mystr)
            back_cam_list.append(rear_camera.group())
        except AttributeError:
            back_cam_list.append('-')

        try:
            battery = re.search(r'[\d]+\s\w{3}\sBattery',mystr)
            battery_list.append(battery.group())
        except AttributeError:
            battery_list.append('-')

        try:
            display_size = re.search(r'[\d]+\D[\d]+\scm',mystr)
            display_list.append(display_size.group())
        except AttributeError:
            display_lists.append('-')

        try:
            processor = re.search(r'Battery\n[\D\w\d\s]+Processor',mystr).group().replace('Battery','')
            processor_list.append(processor.replace('\n',''))
        except:
            processor_list.append('-')
    
    
    #function for getting brand of the phone from phone name
    def brands(x):
        b = x.split()[0]
        return b
    
    
    #below two columns can not be extracted in the first page
    #so lets loop into every page and exatract the data needed
    phone_resolution_list = []
    processor_core_list = []
    
    #since we already have the phone urls, lets loop in
    for i in phone_url_list:
        driver.get(i)
        
        #seeting delay
        delay = 10
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME,'_3dtsli')))
        
        #clicking more button
        more_button = driver.find_element_by_xpath("//button[@class = '_2KpZ6l _1FH0tX']")
        more_button.click()

        try:
            display_resolutions = driver.find_element_by_xpath("//li[contains(text(),\'0 x')]")
            phone_resolution_list.append(display_resolutions.text.replace('$',''))
        except NoSuchElementException:
            phone_resolution_list.append('-')


        try:
            processor_core = driver.find_element_by_xpath("//li[contains(text(),\'Core')]")
            processor_core_list.append(processor_core.text)
        except NoSuchElementException:
            processor_core_list.append('-')  
    

    df_phone_desc = pd.DataFrame({})  #creating the  empty dataframe
    df_phone_desc['Name'] = phone_names_list  
    
    #applying brands function on name column
    df_phone_desc['Brand'] = df_phone_desc['Name'].apply(brands)   
    
    #Creating a column for the particular list
    df_phone_desc['Price'] =price_list
    df_phone_desc['Colour'] =phone_colour_list
    df_phone_desc['RAM']= ram_list
    df_phone_desc['ROM']=rom_list
    df_phone_desc['Front Camera'] =front_cam_list  
    df_phone_desc['Back Camera'] =back_cam_list
    df_phone_desc['Battery Capacity'] =battery_list
    df_phone_desc['Display Size'] =display_list
    df_phone_desc['Display Resolution'] = phone_resolution_list
    df_phone_desc['Processor'] =processor_list
    df_phone_desc['Processor Core'] = processor_core_list
    df_phone_desc['Url'] = phone_url_list
    
    df_phone_desc.to_csv('Phone_df.csv')  #exporting to csv
    
    driver.close() #closing driver
    
    return df_phone_desc    #returning the Dataframe as result
```

## Question 5

```python
def question5(url):
    #activate chrome browser
    driver= webdriver.Chrome("chromedriver.exe")

    #making request
    driver.get(url)  #lets get the google maps
    driver.maximize_window()  #lets maximize our window

    input_cell= driver.find_element_by_id('searchboxinput')
    
    
    #lets create block for user input
    city = input('Enter the city name : ')
    input_cell.send_keys(city)
    
    #searching the city
    search_button = driver.find_element_by_id("searchbox-searchbutton")
    search_button.click()
    
    #lets set the delay based on class
    delay = 15
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME , 'x3AX1-LfntMc-header-title-ma6Yeb-haAclf'))) 
    
    #lets get the url of thhe current webpage
    mystr = driver.current_url
    
    #extracting the latitude and longitude with regex
    a = re.findall(r'\@(-?[\d\.]*,-?[\d\.]*)',mystr)[0]
    
    #assigning it the variables
    latitude = a.split(',')[0]
    longitude = a.split(',')[1]
    
    
    #prininting the result required
    print('\n')
    print(city)
    print('Latitude of {} is : {} '.format(city,latitude))
    print('Longitude of {} is : {}'.format(city,longitude))
    
    driver.close()
```

## Question 6

```python
def question6(url):   
    #activate chrome browser
    driver= webdriver.Chrome("chromedriver.exe")

    #making request
    driver.get(url) 

    delay =10  #10 seconds delay    
    #webdriver wait will wait for 10 seconds
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.ID , 'menu-item-51510')))

    #geeting link of funding deals
    funding_deals = driver.find_element_by_xpath("//li[@id = 'menu-item-51510']/a")
    driver.get(funding_deals.get_attribute('href'))
    driver.maximize_window()  #maximizing window

    delay =10  #10 seconds delay    
    #webdriver wait will wait for 10 seconds
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.ID , 'tablepress-56_wrapper')))
    
    #getting the columns of the dataframe
    heading_list = []
    for i in range(1,10):
        headings_path = "//table[@class = 'tablepress tablepress-id-48 dataTable no-footer']/thead/tr/th[{}]".format(i)
        headings = driver.find_element_by_xpath(headings_path)
        heading_list.append(headings.text)
    df_funding_deals = pd.DataFrame(columns=heading_list, index= list(range(1,31))) #creating the empty dataframe
    
    
    df_funding_deals.loc[1:10,'Month'] = 'July'
    df_funding_deals.loc[11:20,'Month'] = 'August'
    df_funding_deals.loc[21:30,'Month'] = 'Sepetember'
    
    #for three month lets use three for loops
    
    #i tried filling each cell of dataframe individually
    
    # filling row 1 to 10
    for i in range(len(df_funding_deals.columns)):
        for j in range(0,10):
            for k in range(0,9):
                body = driver.find_element_by_xpath("//table[@id = 'tablepress-48']/tbody/tr[{}]/td[{}]".format(j+1,k+1))
                df_funding_deals.iloc[j,k] = body.text
                
    # filling row 10 to 20
    for i in range(len(df_funding_deals.columns)):
        for j in range(10,20):
            for k in range(0,9):
                body = driver.find_element_by_xpath("//table[@id = 'tablepress-49']/tbody/tr[{}]/td[{}]".format((j-10)+1,k+1))
                df_funding_deals.iloc[j,k] = body.text

    # filling row 20 to 30           
    for i in range(len(df_funding_deals.columns)):
        for j in range(20,30):
            for k in range(0,9):
                body = driver.find_element_by_xpath("//table[@id = 'tablepress-50']/tbody/tr[{}]/td[{}]".format((j-20)+1,k+1))
                df_funding_deals.iloc[j,k] = body.text

    driver.close()
    
    return df_funding_deals  #returning the Dataframe as result
```

## Question 7

```python
def question7(url):
    
    #activate chrome browser
    driver= webdriver.Chrome("chromedriver.exe")

    #making request
    driver.get(url) 

    #search button
    search_button = driver.find_element_by_xpath("//div[@class = 'search']")
    #clicking search button
    time.sleep(1)
    search_button.click()
    
    #Finding input cell
    search_input = driver.find_element_by_id("globalPageSearchText")
    
    #sending input as "Top 10 laptops"
    time.sleep(1)
    search_input.send_keys("Top 10 gaming laptops")
    
    #clicking enter button
    time.sleep(1)
    search_input.send_keys(Keys.ENTER)
    
    
    #extracting the links of the top 10 laptops
    top_10_laptops_links = []
    laptops_10 = driver.find_elements_by_xpath("//div[@class = 'searchPage']/a")
    for i in laptops_10:
        top_10_laptops_links.append(i.get_attribute('href'))
    
    
    #creating the empty lists for appending data
    
    name_list = []
    market_status_list = []
    realease_date_list = []
    offical_website_list = []
    price_list = []
    os_list =[]
    display_list = []
    processor_list = []
    
    
    #looping in all the links we just extracted
    
    for link in top_10_laptops_links:
        driver.get(link)  #getting each link at a time
        delay =10  #10 seconds delay
        
        #webdriver wait will wait for 10 seconds
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME , 'heading-wraper'))) 
        
        
        #we will name sepearte try catch for  name, since is name is available in all the pages
        try:
            name = driver.find_element_by_xpath("//div[@class = 'heading-wraper']/h1")
            name_list.append(name.text)
        except NoSuchElementException:
            name_list.append('-')   #if not found we will append '-'
        
        
        #for rest all the columns , we will one try catch loop, 
        #beacuse if one element is not found , none of the elements will be found
        
        try:
            Market_Status = driver.find_element_by_xpath("//div[@class = 'status-bar only-desktop']/div[1]/b")
            market_status_list.append(Market_Status.text)

            Release_Date =  driver.find_element_by_xpath("//div[@class = 'status-bar only-desktop']/div[2]/b")
            realease_date_list.append(Release_Date.text)

            Offical_Website = driver.find_element_by_xpath("//div[@class = 'status-bar only-desktop']/div[3]/b")
            offical_website_list.append(Offical_Website.text)

            Price = driver.find_element_by_xpath("//div[@class = 'Block-price']/b")
            price_list.append(Price.text)

            OS  = driver.find_element_by_xpath("//div[@class = 'Specs-Wrap  notspecs ']/ul/li[1]/div")
            os_list.append(OS.text)

            Display  = driver.find_element_by_xpath("//div[@class = 'Specs-Wrap  notspecs ']/ul/li[2]/div")
            display_list.append(Display.text)

            Processor  = driver.find_element_by_xpath("//div[@class = 'Specs-Wrap  notspecs ']/ul/li[3]/div")
            processor_list.append(Processor.text)

        except NoSuchElementException:

            market_status_list.append('-')  #if not found we will append '-' 
            realease_date_list.append('-')
            offical_website_list.append('-')
            price_list.append('-')
            os_list.append('-')
            display_list.append('-')
            processor_list.append('-')

        top_10_gaming_laptop_df = pd.DataFrame({})    #creating a empty dataframe
        
        top_10_gaming_laptop_df['Name'] = name_list    #Creating a column for the particular list
        top_10_gaming_laptop_df['Market Status'] = market_status_list
        top_10_gaming_laptop_df['Release Date'] =realease_date_list
        top_10_gaming_laptop_df['Price'] = price_list
        top_10_gaming_laptop_df['Offical Website'] =offical_website_list
        top_10_gaming_laptop_df['Operating System'] = os_list
        top_10_gaming_laptop_df['Display'] = display_list
        top_10_gaming_laptop_df['Processor'] = processor_list
    
    driver.close()
    return top_10_gaming_laptop_df     #returning the Dataframe as result
```
