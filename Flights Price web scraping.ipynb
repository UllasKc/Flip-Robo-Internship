```python
#lets import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import re
pd.set_option('display.max_row',None)
```

```python
driver = webdriver.Chrome('chromedriver.exe')  #activating chrome browzer
driver.get("https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=CCU&destinationCountry=IN&flexi=0&flight_depart_date=0{}%2F10%2F2021&hb=0&noOfSegments=1&origin=BOM&originCountry=IN&type=O&unique=435896930836&version=1.1&viewName=normal")  #getting url
driver.maximize_window() #maximizing window
```

```python
airline_list = []
date_list= []
source_list =[]
destination_list =[]
depature_list = []
arrival_list = []
total_time_list = []
stop_list =[]
price_list =[]



for date in range(9,32):
    if len(str(date)) == 1:
        date = "0"+str(date)
    try:  
        driver.get("https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=CCU&destinationCountry=IN&flexi=0&flight_depart_date=0{}%2F10%2F2021&hb=0&noOfSegments=1&origin=BOM&originCountry=IN&type=O&unique=435896930836&version=1.1&viewName=normal".format(date))
        delay = 40
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"//div[@class='i-b tipsy fare-summary-tooltip fs-18']")))
    except:
        #driver = webdriver.Chrome('chromedriver.exe')  #activating chrome browzer
        continue

    #flag = driver.find_element_by_xpath("//div[@class='copyright js_copyright_hide']")
    #driver.execute_script("arguments[0].scrollIntoView();",flag)
    for i in range(4):
        driver.execute_script("window.scrollBy(0,document.body.scrollHeight)")
        time.sleep(2)
        
    airline = driver.find_elements_by_xpath("//div[@class='fs-13 airline-name no-pad col-8']/span")
    for i in airline:
        airline_list.append(i.text)
        date_list.append(str(date)+"-10-2021")
             
    source = driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p")
    for i in source:
        source_list.append(i.text)
    
    destination = driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]")
    for i in destination:
        destination_list.append(i.text)
    
    depature = driver.find_elements_by_xpath("//div[@class='fs-15 bold time']")
    for i in depature:
        depature_list.append(i.text)
        
    arrival = driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']")
    for i in arrival:
        arrival_list.append(i.text)
    
    total_time = driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']")
    for i in total_time:
        total_time_list.append(i.text)
       
    stop = driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']/span")
    for i in stop:
        stop_list.append(i.text)
        
    price = driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")
    for i in price:
        price_list.append(i.text)
```

```python
flight_df = pd.DataFrame({})

flight_df["Airline"]= airline_list 
flight_df["Date"]= date_list
flight_df["Source City"]= source_list 
flight_df["Destination City"]= destination_list
flight_df["Departure Time"]= depature_list 
flight_df["Arrival Time"]= arrival_list 
flight_df["Duration"]= total_time_list 
flight_df["Stops"]= stop_list 
flight_df["Price"]= price_list
```
